```python
import os
import pandas
import matplotlib.pyplot as plt
from datetime import datetime  # for using dates and time
from pytz import utc  # as time zone or system used in 'Timestamps' is utc.so for setting timesystem, it is imported.

data = pandas.read_csv("reviews.csv" , parse_dates = ['Timestamp'])
data
```

## Daywise analysis

```python
data["Day"] = data["Timestamp"].dt.date
day_avg = data.groupby(["Day"]).mean()
day_avg
# dt is a feature using which we can get date , month , yr ,etc
# dt.date gives date fromm  timestamp
#groupby() works by grouping data in a group having a same groupby specified format.
#data.groupby(["Day"]).mean()  gives mean rating datewise
```

```python
day_count = data.groupby(["Day"]).count()
day_count
# it gives count of rating and comments per day
```

```python
day_avg
# here Day is index not column
```

```python
plt.plot(day_avg.index , day_avg["Rating"])
#plt.plot(x,y) , x takes the data to be ob x axis and y on y axis
# matplotlib takes range of max and min value automatically from dataset
# but disadvantage of it is that its plot is not interactive
```

```python
plt.figure(figsize=(25,7))
plt.plot(day_avg.index , day_avg["Rating"])
#In Matplotlib all the diagrams are created at a default size of 6.4 x 4.8 inches
#figsize =(widh , height) , used to change dimensions of the plot
```

## Downsampling data 

### on weekly basis

```python
data["Week"] = data["Timestamp"].dt.week
data
# here it gives aggregated weeks i.e. 1st week of 2018,2019 and 2020 are considered as 1st week.
# similarly all weeks are aggregated
```

```python
data["Week"] = data["Timestamp"].dt.isocalendar().week
data
# it is again giving the same as above code i.e weeks are getting aggregated
```

```python
data["Week"] = data["Timestamp"].dt.strftime("%Y-%U")
data
# this time we get yearwise week
# %Y = year and %U = week
# strftime("%Y-%U") gives and strftime = string from time i.e time is obtained from str
```

```python
week_avg = data.groupby(["Week"]).mean()
week_avg
```

```python
plt.plot(week_avg.index , week_avg["Rating"])
```

```python
plt.figure(figsize=(16,8))
plt.plot(week_avg.index , week_avg["Rating"])
```

```python
data["Weekday"] = data['Timestamp'].dt.strftime("%A")
data["daynum"] = data["Timestamp"].dt.strftime('%w') # we added this as sorting would have been alphabetical due to str
weekday_avg1 = data.groupby(['Weekday','daynum']).mean()
weekday_avg1 = weekday_avg1.sort_values('daynum')# sorts the data on daynumber basis
weekday_avg1
```

```python
weekday_avg1.plot(figsize = (20,4))
```

```python
# we can also use this
data["Weekday"] = data['Timestamp'].dt.strftime("%A")
data["daynum"] = data["Timestamp"].dt.strftime('%w')
weekday_avg1 = data.groupby(['Weekday','daynum']).mean()
weekday_avg1 = weekday_avg1.sort_values('daynum')

plt.figure(figsize=(20,4))
plt.plot(weekday_avg1.index.get_level_values(0) , weekday_avg1['Rating'])
```
